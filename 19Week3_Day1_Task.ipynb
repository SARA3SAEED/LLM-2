```python
from datasets import load_dataset

ds = load_dataset("salehalmansour/english-to-arabic-translate")
```

```python
import pandas as pd

data = ds['train']
df = pd.DataFrame(data)
df.head()
```

```python
import pandas as pd


new_data = {'Prompt': [], 'Response': []}


for i in range(0, len(df), 4):
    if i + 3 < len(df):
        prompt = "Translate the following English text to Arabic:\n"
        for j in range(i, i+3):
            english_text = df.loc[j, 'en']
            arabic_text = df.loc[j, 'ar']
            prompt += f"{english_text}\n\"{arabic_text}\"\n"

        english_text = df.loc[i+3, 'en']
        prompt += f"{english_text}\n"

        response = f"\"{df.loc[i+3, 'ar']}\""


        new_data['Prompt'].append(prompt.strip())
        new_data['Response'].append(response)


df_new = pd.DataFrame(new_data)


print(df_new.head())
```
