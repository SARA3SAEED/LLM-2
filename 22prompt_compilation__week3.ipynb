#First_Tamplate

```python
import pandas as pd

def create_prompt_and_compilation(df, shot=0):
    if len(df) < 1:
        raise ValueError("DataFrame must contain at least 1 row")

    if shot < 0:
        raise ValueError("Shot must be a non-negative integer")

    Q_A_templates = [
        "السؤال: {question}?\nالإجابة: {answer}",
        "الرجاء الاجابة على هذا السؤال: {question}, {answer}"
    ]

    prompts = []
    compilations = []

    for template in Q_A_templates:
        if shot == 0:

            for i in range(len(df)):
                prompt = template.format(question=df.iloc[i]['question'], answer='')
                compilation = df.iloc[i]['answer']
                prompts.append(prompt)
                compilations.append(compilation)
        else:

            step = shot + 1
            total_rows = len(df)
            for i in range(0, total_rows - step + 1, step):
                prompt = ""
                for j in range(step - 1):
                    if i + j < total_rows:
                        prompt += template.format(question=df.iloc[i + j]['question'], answer=df.iloc[i + j]['answer']) + "\n"


                prompt += template.format(question=df.iloc[i + shot]['question'], answer='')


                compilation = df.iloc[i + shot]['answer']

                prompts.append(prompt)
                compilations.append(compilation)

    result_df = pd.DataFrame({
        'prompt': prompts,
        'compilation': compilations
    })

    return result_df
```

```python
import pandas as pd

def create_prompt_and_compilation_2(df, shot=0):
    if len(df) < 1:
        raise ValueError("DataFrame must contain at least 1 row")

    if shot < 0:
        raise ValueError("Shot must be a non-negative integer")

    templates = [
        "النص: {source}?\n التلخيص: {target}",
        "الرجاء تلخيص هذا النص: {source}, {target}"
    ]

    prompts = []
    compilations = []

    for template in templates:
        if shot == 0:

            for i in range(len(df)):
                prompt = template.format(source=df.iloc[i]['source'], target='')
                compilation = df.iloc[i]['target']
                prompts.append(prompt)
                compilations.append(compilation)
        else:

            step = shot + 1
            total_rows = len(df)
            for i in range(0, total_rows - step + 1, step):
                prompt = ""
                for j in range(step - 1):
                    if i + j < total_rows:
                        prompt += template.format(source=df.iloc[i + j]['source'], target=df.iloc[i + j]['target']) + "\n"


                prompt += template.format(source=df.iloc[i + shot]['source'], target='')


                compilation = df.iloc[i + shot]['target']

                prompts.append(prompt)
                compilations.append(compilation)

    result_df = pd.DataFrame({
        'prompt': prompts,
        'compilation': compilations
    })

    return result_df
```
