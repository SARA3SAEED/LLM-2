```python
from datasets import load_dataset

ds = load_dataset("nazimali/quran-question-answer-context")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
```

```python
import pandas as pd
dataset = ds['train']
df = pd.DataFrame(dataset)
df.head()
```

```python
import random
```

```python
def generate_prompt_completion_dataset(df, num_shots=0, mode="zero-shot"):
    templates = [
        "Question: {question}?\nAnswer:",
        "{question}?",
        "Answer the following question:\n\n{question}",
        "Answer this question:\n\n{question}?",
        "Please answer this question: {question}",
        "Answer the question...{question}?",
        "What is the answer to this question? {question}\n\n",
        "Can you tell me the answer to {question}?",
        "Next question: {question}\n\n",
        "Q: {question} A:"
    ]

    data = []

    for i in range(len(df)):
        template = random.choice(templates)

        if mode == "zero-shot":
            prompt = template.format(question=df.iloc[i]['question'])
            completion = df.iloc[i]['answer']
            data.append({"prompt": prompt, "completion": completion})

        elif mode == "few-shot":

            if i >= num_shots:
                shot_prompts = []
                for j in range(i - num_shots, i):
                    shot_prompts.append(f"Q: {df.iloc[j]['question']} A: {df.iloc[j]['answer']}")
                shot_prompt = "\n\n".join(shot_prompts)


                prompt = f"{shot_prompt}\n\n{template.format(question=df.iloc[i]['question'])}"
                completion = df.iloc[i]['answer']
                data.append({"prompt": prompt, "completion": completion})
            else:
                continue

    return pd.DataFrame(data)
```

```python
df_zero_shot = generate_prompt_completion_dataset(df, mode="zero-shot")
```

```python
from datasets import load_dataset

ds_sum = load_dataset("Abdelkareem/wikihow-arabic-summarization")
```

```python
import pandas as pd
dataset_sum = ds_sum['train']
df_sum = pd.DataFrame(dataset_sum)
df_sum.head()
```

```python
def generate_summarization_dataset(df, num_shots=0, mode="zero-shot"):
    templates = [
        "Write a short summary for this text: {text}\n\nSummary:",
        "Briefly summarize this sentence: {text}\n\nSummary:",
        "Generate a short summary of this sentence:\n{text}\n\nSummary:",
        "What is a shorter version of this:\n\n{text}\n\nSummary:",
        "{text}\n\nWrite a brief summary in a sentence or less.",
        "{text}\n\nWhat is a very short summary of the above text?",
        "{text}\nSummarize the aforementioned text in a single phrase.",
        "{text}\nCan you generate a short summary of the above paragraph?",
        "Write a text based on this summary: {summary}\n\nText:",
        "Write a text based on \"{summary}\"\n\nText:"
    ]

    data = []

    for i in range(len(df)):
        template = random.choice(templates)

        if mode == "zero-shot":
            if "{text}" in template:
                prompt = template.format(text=df.iloc[i]['text'])
                completion = df.iloc[i]['summary']
            else:
                prompt = template.format(summary=df.iloc[i]['summary'])
                completion = df.iloc[i]['text']
            data.append({"prompt": prompt, "completion": completion})

        elif mode == "few-shot":
            # Collect previous shots
            if i >= num_shots:
                shot_prompts = []
                for j in range(i - num_shots, i):
                    if "{text}" in template:
                        shot_prompts.append(f"Text: {df.iloc[j]['text']}\nSummary: {df.iloc[j]['summary']}")
                    else:
                        shot_prompts.append(f"Summary: {df.iloc[j]['summary']}\nText: {df.iloc[j]['text']}")

                shot_prompt = "\n\n".join(shot_prompts)


                if "{text}" in template:
                    prompt = f"{shot_prompt}\n\n{template.format(text=df.iloc[i]['text'])}"
                    completion = df.iloc[i]['summary']
                else:
                    prompt = f"{shot_prompt}\n\n{template.format(summary=df.iloc[i]['summary'])}"
                    completion = df.iloc[i]['text']
                data.append({"prompt": prompt, "completion": completion})
            else:
                continue

    return pd.DataFrame(data)
```

```python
df_zero_shot_sum = generate_summarization_dataset(df_sum, mode="zero-shot")
df_zero_shot_sum.head()
```

```python
df_few_shot_sum = generate_summarization_dataset(df_sum, 2,mode="few-shot")
df_few_shot_sum.head()
```
