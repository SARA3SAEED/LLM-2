```python
import random
import string
import numpy as np
import pandas as pd
from datasets import load_dataset
```

```python
# Source of dataset: https://huggingface.co/datasets/Harvard/gigaword
dataset = load_dataset("Harvard/gigaword", split = "test", trust_remote_code=True)
df = dataset.to_pandas()
```

```python
df.rename(columns = {'document':'text'}, inplace = True)
df.head(5)
```

```python
# Source of templates: https://github.com/google-research/FLAN/blob/main/flan/templates.py
templates = [
    ("Summarize the following text in one or two sentences:\n{text}\n{summary}"),
    ("Provide a brief summary for this sentence:\n{text}\n{summary}"),
    ("Generate a concise summary for this sentence:\n{text}\n{summary}"),
    ("What is a concise summary of the following text?\n{text}\n{summary}"),
    ("{text}\nPlease summarize the above text in one sentence.\n{summary}"),
    ("{text}\nCan you provide a very short summary of the text above?\n{summary}"),
    ("{text}\nSummarize the above text in a single, clear phrase.\n{summary}"),
    ("{text}\nCan you create a brief summary of the paragraph provided?\n{summary}"),
    ("Compose a sentence based on this summary:\n{summary}\n{text}"),
    ("Craft a sentence that reflects the following summary:\n\"{summary}\"\n{text}"),
]
```

```python
def generate_prompts(query, number):
    if number > 0:
        number -= 1

    selected_templates = random.choices(templates, k=number)
    selected_rows = df.sample(n=number)

    for template, (_, row) in zip(selected_templates, selected_rows.iterrows()):
        prompt = template.format(text=row['document'], summary=row['summary'])
        print(prompt)
        print("===================")


    user_prompt = f"Provide a brief summary for the following text: {query}\n"
    print(user_prompt)
```

```python
# Example usage
# For zero-shot, simply set number to zero. For few-shot, set it to a positive number.
query = "Explain the impact of global warming"
number = 0
generate_prompts(query, number)
```
